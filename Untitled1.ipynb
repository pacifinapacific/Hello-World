```python


class Pix2PixPairFolder(ImageFolder):
    def __init__(self, root):
        super().__init__(root)
        self.resizer=transforms.Scale(300,600)
        self.transform=transforms.ToTensor()
    def __gettim__(self,index):
        path, _=self.imgs[index]
        img=self.loader(path)
            
        img_r=self.resizer(img)
        img_r=self.transform(img_r)
        photo=img_r[:,:,:300]
        mask=img_r[:,:,300:]

        
        return img_r,photo,mask
        
        
```
